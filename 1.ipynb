```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
eps0 = 8.85e-12
e = 1.6e-19
p = 1e5
NE0 = 1e18
E0 = 5e3
F0 = 1500
U1 = -100
U2 = 0
mue = 1
mui = 1
alpha = 1
ne_0 = 1e18
k = 1.380649e-23
T = 300

# Grid parameters
N = 50
M = N
L = 1
R = 1
hr = 1 / (N - 0.5)
hz = L / (M - 1)

# Initialize arrays
ne = np.zeros((N, M))
ni = np.zeros((N, M))
r = np.linspace(hr / 2, R, N)
z = np.linspace(0, L, M)

# Calculate ne
for j in range(M):
    for i in range(N):
        ne[i, j] = ne_0 * (1 - (r[i] / R) ** 2) * (z[j] * (L - z[j]))

for j in range(M):
    for i in range(N):
        if r[i] < 0.85:
            ni[i, j] = ne[i, j]
        else:
            ni[i, j] = ne[i, j] + (r[i] - 0.85) / (R - 0.85) * (0.3e17 - ne[i, j])
            ni[i, j] = max(ne[i, j], ni[i, j])

plt.figure(figsize=(8, 6))
plt.plot(r, ne[:, M//2], label='ne')
plt.plot(r, ni[:, M//2], label='ni')
plt.xlabel('r')
plt.grid(True)
plt.legend()
plt.show()
```
